Now we move for Numerical colums


Not always — but you should at least use one visualization for each variable type to understand its pattern.

Here’s a good rule of thumb 👇


Variable Type	Minimum Required	Optional/Extra

Categorical	Count plot	Pie chart


Numerical	Histogram or Box plot	Violin plot, KDE plot

🔹 Summary:
Step	Variable Type	Tools	Purpose

1	Categorical	Count plot	See frequency of categories

2	Numerical	Histogram	Understand data shape


3	Numerical	Box plot	Detect outliers, spread


4	(Optional)	Violin / KDE plot	Get richer view of distribution
