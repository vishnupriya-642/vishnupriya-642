```python
from sentence_transformers import SentenceTransformer, util
import pdfplumber
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

```

```python
resume_text = "I have 2 years of experience in Python, machine learning, NLP and data analytics."
job_description = "Looking for a machine learning engineer with strong Python and NLP experience."

embedding_resume = model.encode(resume_text)
embedding_jd = model.encode(job_description)

score = util.cos_sim(embedding_resume, embedding_jd).item()
print("Similarity Score:", round(score, 2))
```
